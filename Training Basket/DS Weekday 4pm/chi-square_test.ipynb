```python
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_wine
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['Cabin'], inplace = True)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
categorical_data = df[['Sex', 'SibSp', 'Parch', 'Embarked']].dropna()
```

```python
categorical_data.head()
```

```python
le = LabelEncoder()
```

```python
le
```

```python
for col in categorical_data.columns:
    categorical_data[col] = le.fit_transform(categorical_data[col])
```

```python
categorical_data.head(10)
```

```python
y = df['Survived'].dropna()
```

```python
y
```

```python
X = categorical_data
```

```python
chi_statistics , p_value = chi2(X, y)
```

```python
chi_statistics
```

```python
p_value
```

```python
for i in range(len(p_value)):
    if p_value[i] < 0.05:
        print(categorical_data.columns[i])
```
