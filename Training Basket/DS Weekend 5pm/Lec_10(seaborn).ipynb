```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Line plot
# first graph
plt.subplot(1, 2, 1)  # row 1 and column 2 1st graph 
sns.lineplot(x = df['sepal_length'], y = df['sepal_width'] , hue = df['species'] )
plt.title("Line plot sepal length and sepal width")
plt.xticks(np.arange(df['sepal_length'].min(), df['sepal_length'].max(), step = 1))
plt.yticks(np.arange(df['petal_width'].min(), df['petal_width'].max(), step = 1))

#second graph
plt.subplot(1, 2, 2)  # row 1 and column 2 2st graph 
sns.lineplot(x = df['petal_length'], y = df['petal_width'] , hue = df['species'] )
plt.xticks(np.arange(df['petal_length'].min(), df['petal_length'].max(), step = 1))
plt.yticks(np.arange(df['petal_width'].min(), df['petal_width'].max(), step = 1))
plt.title("Line plot petal length and petal width")

plt.tight_layout()
plt.show()
```

```python
# Scatter plot
# Line plot
# first graph
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)  # row 1 and column 2 1st graph 
sns.scatterplot(x = df['sepal_length'], y = df['sepal_width'] , hue = df['species'] )
plt.title("scatter plot sepal length and sepal width")
plt.xticks(np.arange(df['sepal_length'].min(), df['sepal_length'].max(), step = 1))
plt.yticks(np.arange(df['petal_width'].min(), df['petal_width'].max(), step = 1))

#second graph
plt.subplot(1, 2, 2)  # row 1 and column 2 2st graph 
sns.scatterplot(x = df['petal_length'], y = df['petal_width'] , hue = df['species'] )
plt.xticks(np.arange(df['petal_length'].min(), df['petal_length'].max(), step = 1))
plt.yticks(np.arange(df['petal_width'].min(), df['petal_width'].max(), step = 1))
plt.title("scatter plot petal length and petal width")

plt.tight_layout()
plt.show()
```

```python
# bar graph
plt.figure(figsize=(12, 5))

# 1st graph
plt.subplot(2, 2, 1) 
sns.barplot(x = df['species'], y = df['sepal_length'], hue = df['species'])
plt.title('Sepal length of all Categories')
plt.yticks(np.arange(df['sepal_length'].min(), df['sepal_length'].max(), step = 2))
plt.xticks(rotation = 45)
plt.grid()

#2nd graph
plt.subplot(2, 2, 2) 
sns.barplot(x = df['species'], y = df['sepal_width'], hue = df['species'])
plt.title('Sepal Width of all Categories')
plt.yticks(np.arange(df['sepal_width'].min(), df['sepal_width'].max(), step = 2))
plt.xticks(rotation = 45)
plt.grid()

#3rd graph
plt.subplot(2, 2, 3) 
sns.barplot(x = df['species'], y = df['petal_length'], hue = df['species'])
plt.title('petal length of all Categories')
plt.yticks(np.arange(df['petal_length'].min(), df['petal_length'].max(), step = 2))
plt.xticks(rotation = 45)
plt.grid()

#4 graph
plt.subplot(2, 2, 4) 
sns.barplot(x = df['species'], y = df['petal_width'], hue = df['species'])
plt.title('petal_width of all Categories')
plt.yticks(np.arange(df['petal_width'].min(), df['petal_width'].max(), step = 2))
plt.xticks(rotation = 45)
plt.grid

plt.tight_layout()
plt.show()
```

```python
# countplot
plt.figure(figsize=(12, 5))

sns.countplot(x = df['species'], hue = df['species'])
plt.xticks(rotation = 90)
plt.show()
```

```python
# Box plot 
plt.figure(figsize=(12, 5))
plt.subplot(2, 2, 1)
sns.boxplot(x = df['species'], y = df['sepal_length'], color = 'g', hue = df['species'])
plt.title('sepal_length of all Categories')
plt.xticks(rotation = 45)
plt.grid()

plt.subplot(2, 2, 2)
sns.boxplot(x = df['species'], y = df['sepal_width'], color = 'r',  hue = df['species'])
plt.title('sepal_width of all Categories')
plt.xticks(rotation = 45)
plt.grid()

plt.subplot(2, 2, 3)
sns.boxplot(x = df['species'], y = df['petal_length'], color = 'y',  hue = df['species'])
plt.title('petal_length of all Categories')
plt.xticks(rotation = 45)
plt.grid()

plt.subplot(2, 2, 4)
sns.boxplot(x = df['species'], y = df['petal_width'], color = 'b',  hue = df['species'])
plt.title('petal_width of all Categories')
plt.xticks(rotation = 45)
plt.grid()

plt.tight_layout()
```
