```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot : when we want to understand the relationship between two numerical columns or features

```python
x = np.arange(0, 100) # independent variable
y = 2*x + 3  # dependent variable
```

```python
len(y)
```

```python
len(x)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, y, color = 'r', linestyle = '--', label = 'X-Y graph')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear Relationship')
plt.show()
```

```python
x1 = np.random.randint(0, 100, 50)
```

```python
len(x1)
```

```python
y1 = 3*x1 + 3
len(y1)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x1, y1,color = 'y', label = 'X1-Y1 graph')
plt.plot(x, y, color = 'r', linestyle = '--', label = 'X-Y graph')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear Relationship')
plt.show()
```

# Scatter plot : when we want to understand the relationship between two numerical columns or features by showing you the points

```python
plt.figure(figsize=(10, 4))
plt.scatter(x1, y1,color = 'y', label = 'X1-Y1 graph')
plt.scatter(x, y, color = 'r', marker = '^' , label = 'X-Y graph')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear Relationship')
plt.show()
```

# Bar graph : to analyze the categorical variables 

```python
students = ['Tushar', 'Indu','Rachit']
heights = [160, 150, 170]
```

```python
plt.bar(students, height=heights, width= 0.5, color = 'r', label = 'Bar Graph')
plt.legend()
plt.xlabel('Students name')
plt.ylabel('Heights')
plt.title('Height measurements')
plt.show()
```

# histogram : to understand the distribution of any numerical variable 

```python
x = np.random.randn(50)
```

```python
x
```

```python
plt.hist(x, bins = 20, label = 'Histogram', color='m')
plt.legend()
plt.title('Distribution of X')
plt.show()
```

# Pie Chart : 

```python
plt.pie(heights, labels=students, autopct='%.2f', explode=[0, 0.1, 0], shadow=True)
plt.show()
```
