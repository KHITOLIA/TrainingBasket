```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Binomial Distribution 
n = 10  # number of trials
p = 0.9 # tossing a coin
num_sample = 1000
sample_size = 10
binomial_dist = np.random.binomial(n , p , size = (num_sample, sample_size))
```

```python
binomial_dist[0]
```

```python
binomial_dist.mean()
```

```python
for i in range(2):
    plt.hist(binomial_dist[i], density=True, color = 'red', edgecolor = 'yellow')
    plt.title("Histogram of Binomial Dist")
    plt.xlabel('number of trials')
    plt.ylabel('Probability')
    plt.show()
```

```python
binomial_mean = np.mean(binomial_dist, axis = 1)
```

```python
plt.hist(binomial_mean, density=True, color = 'red', edgecolor = 'yellow')
plt.title("Histogram of Binomial Dist")
plt.xlabel('number of trials')
plt.ylabel('Probability')
plt.show()
```

```python
binomial_mean.mean()
```

```python
# Poisson Distribution 

num_sample = 1000
sample_size = 50
lambda_param = 5

theoretical_mean = lambda_param

poisson_dist = np.random.poisson(lam = lambda_param, size = (num_sample, sample_size))
```

```python
poisson_dist
```

```python
poisson_dist.mean()
```

```python
poisson_mean = np.mean(poisson_dist, axis = 1)
plt.hist(poisson_mean, density=True, color = 'red', edgecolor = 'yellow')
plt.show()
```

```python
poisson_mean.mean().round() == lambda_param
```

```python
# Exponential distribution

lambda_param = 2
num_sample = 1000
sample_size = 80

theoretical_mean = 1/lambda_param
expo_dist = np.random.exponential(scale = lambda_param, size = (num_sample, sample_size))
```

```python
expo_dist
```

```python
expo_dist.mean()
```

```python
theoretical_mean
```

```python
expo_mean = np.mean(expo_dist, axis = 1)
```

```python
plt.hist(expo_mean, density=True, color = 'red', edgecolor = 'yellow')
plt.show()
```
