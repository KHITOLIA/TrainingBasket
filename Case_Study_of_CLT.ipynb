# Uniform distribution

# Exponential Distribution

```python
num_samples = 50000
sample_size = 500
lambda_param = 2

samples_expo = np.random.exponential(scale = 1/lambda_param, size = (num_samples, sample_size))
```

```python
sample_means_expo = np.mean(samples_expo, axis = 1)
```

```python
plt.hist(sample_means_expo, bins = 40, density = True, edgecolor = 'yellow')
plt.title("Histogram of Sample means(Exponential Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

# POISSON DISTRIBUTION
# GAMMA DISTRIBUTION
# BINOMIAL DISTRIBUTION

```python
num_samples = 1000
samnple_size = 50

poisson_lambda = 5

gamma_shape = 2
gamma_scale = 1

binomial_n = 10
binomial_p = 0.5

poisson_sample = np.random.poisson(poisson_lambda, size = (num_samples, sample_size))
gamma_sample = np.random.gamma(gamma_shape, scale = gamma_scale, size = (num_samples, sample_size))
binomial_sample = np.random.binomial(binomial_n, binomial_p, size = (num_samples, sample_size))
```

```python
# Mean of each samples
```

```python
poisson_means = np.mean(poisson_sample, axis = 1)
gamma_means = np.mean(gamma_sample, axis = 1)
binomial_means = np.mean(binomial_sample, axis = 1)
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(3, 1, 1)
plt.hist(poisson_means, bins = 40, density = True, edgecolor = 'yellow')
plt.title("Histogram of Sample means(Poisson Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.subplot(3, 1, 2)
plt.hist(gamma_means, bins = 40, density = True, edgecolor = 'yellow')
plt.title("Histogram of Sample means(Gamma Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.subplot(3, 1, 3)
plt.hist(binomial_means, bins = 40, density = True, edgecolor = 'yellow')
plt.title("Histogram of Sample means(Binomial Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
population_size = 100000
sample_size = 50
num_samples = 100

# Generate a random representative sample of salaries(in thousands)
# you should replace this with the collected salary data
population_salaries = np.random.lognormal(mean = 4, sigma = 0.8, size = population_size)
```

```python
population_salaries
```

```python
plt.hist(population_salaries)
```

```python
np.random.choice(population_salaries)  # it is in numpy format that's why used as choice
```

```python
sample_mean = []
sample_std = []

for i in range(2):
    sample_salaries = np.random.choice(population_salaries, size = sample_size)
    sample_mean.append(np.mean(sample_salaries))
    sample_std.append(np.std(sample_salaries))
```

```python
# Calculate the average of the sample means and standard error
average_sample_means = np.mean(sample_mean)
standard_error = np.std(sample_mean) / np.sqrt(num_samples)
```

```python
# Calculate the 95% confidence interval
margin_of_error = 1.96 * standard_error
lower_limit = average_sample_means - margin_of_error
upper_limit = average_sample_means + margin_of_error
```

```python
# Report the results
print(f"Estimated average salary (in thousands): {average_sample_means:.2f}")
print(f"95% confidence interval (in thousands): ({lower_limit:.2f}, {upper_limit:.2f})")
```
